```python
feature_name=list(df.columns.values)
feature_name=feature_name[1:]
feature_name=feature_name[:-1]
print(feature_name)
len(feature_name)
```

```python
input_data=df.values[:,1:]
input_data
```

```python
set(list(input_data[:,17]))
```

```python
# convert to catigorical variable 
continues_index=[0,3,10] # continues feature, skip
for i in range(input_data.shape[1]):
    if i in continues_index:
        continue
    dic_lable={}
    count=0
    for j in range(input_data.shape[0]):
        if input_data[j,i] not in dic_lable:
            dic_lable[input_data[j,i]]=count
            count=count+1
        else:
            continue
    for j in range(input_data.shape[0]):
        input_data[j,i]=dic_lable[input_data[j,i]]
```

```python
No_Fraud_data,Fraud_data = [], []
No_Fraud_index,Fraud_index=[], []
for count, elem in enumerate(input_data):
    if elem[-1] == 0:
        No_Fraud_data.append(elem.tolist())
        No_Fraud_index.append(count)
    else:
        Fraud_data.append(elem.tolist())
        Fraud_index.append(count)
```

```python
import numpy as np
Train_data = np.array(No_Fraud_data[:-50]+Fraud_data[:-50])
Test_data  = np.array(No_Fraud_data[-50:]+Fraud_data[-50:])
```

```python
# upsampling "fraud data"
import numpy as np
Train_data_temp=list(Train_data).copy()
ratio=sum(Train_data[:,-1])/len(Train_data[:,-1])
for elem in Train_data:
    if elem[-1] == 1:
        for i in range(int(np.ceil(1/ratio))):
            Train_data_temp.append(elem)

Train_data=np.array(Train_data_temp)
```

```python
# fit tree model
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
# X = input_data[:, :-1]
# Y = input_data[:,-1]
# X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size = 0.3, random_state = 100)

X_train,y_train=Train_data[:,:-1],Train_data[:,-1]
X_test, y_test =Test_data[:,:-1], Test_data[:,-1]

y_train=y_train.astype('int')
y_test =y_test.astype('int')

clf_gini = DecisionTreeClassifier(criterion = "gini", random_state = 100,
                               max_depth=5, min_samples_leaf=100)
clf_gini.fit(X_train, y_train)
```

```python
import collections
import pydotplus
dot_data = tree.export_graphviz(clf_gini,
                                feature_names=feature_name,
                                out_file=None,
                                filled=True,
                                rounded=True)
# graph = pydotplus.graph_from_dot_data(dot_data)
# colors = ('turquoise', 'orange')
# edges = collections.defaultdict(list)

# for edge in graph.get_edge_list():
#     edges[edge.get_source()].append(int(edge.get_destination()))

# for edge in edges:
#     edges[edge].sort()    
#     for i in range(2):
#         dest = graph.get_node(str(edges[edge][i]))[0]
#         dest.set_fillcolor(colors[i])

# graph.write_png('tree.png')
```

```python
test_ratio=len(y_test[y_test==1])/len(y_test)
print('fraud/non-fraud ratio in testing data is:',test_ratio)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.metrics import f1_score
y_pred = clf_gini.predict(X_test)
y_pred_train = clf_gini.predict(X_train)
print("Traing accuracy is {}%".format(accuracy_score(y_train,y_pred_train)*100))
print("Testing accuracy is {}%".format(accuracy_score(y_test,y_pred)*100))
print('confusion matrix:')
print(confusion_matrix(y_test,y_pred))
print('F1 score:')
print(f1_score(y_test,y_pred))
```

```python
# convert to catigorical variable 
continues_index=[0,3,10] # continues feature, skip
feature_size=0
for i in range(input_data.shape[1]):
    if i in continues_index:
        continue
    else:
        print('feature num:{},column:{}'.format(max(input_data[:,i]),i))
        feature_size=max(input_data[:,i])+feature_size+1
        
feature_size=feature_size+len(continues_index)
input_data_onehot=np.zeros([len(input_data),feature_size])
# update the data matrix
count=0
for i in range(input_data.shape[1]):
    if i in continues_index:
        print('column {}'.format(count))
        input_data_onehot[:,count]=input_data[:,i]
        count=count+1
    else:
        print('column {} to {}'.format(count,(count+max(input_data[:,i])+1)))
        box=np.zeros([len(input_data[:,i]),max(input_data[:,i])+1])
        box[np.arange(box.shape[0]),input_data[:,i].astype(int)] = 1
        input_data_onehot[:,count:(count+max(input_data[:,i])+1)]=box
        count=count+max(input_data[:,i])+1

input_data_onehot=input_data_onehot[:,:-1]
input_data_onehot[:,-1]=abs(input_data_onehot[:,-1]-1) # just reverse the label
```

```python
Fraud_data_onehot = input_data_onehot[Fraud_index]
Non_Fraud_data_onehot=input_data_onehot[No_Fraud_index]

Train_data_onehot = np.array(Non_Fraud_data_onehot[:-50].tolist()+Fraud_data_onehot[:-50].tolist())
Test_data_onehot  = np.array(Non_Fraud_data_onehot[-50:].tolist()+Fraud_data_onehot[-50:].tolist())

# upsampling "fraud data"
Train_data_temp=list(Train_data_onehot).copy()
ratio=sum(Train_data_onehot[:,-1])/len(Train_data_onehot[:,-1])
print('ratio:',ratio)
for elem in Train_data_onehot:
    if elem[-1] == 1:
        for i in range(int(np.ceil(1/ratio))):
            Train_data_temp.append(elem)

Train_data_onehot=np.array(Train_data_temp)

X_train_onehot, y_train_onehot= Train_data_onehot[:,:-1],Train_data_onehot[:,-1]
X_test_onehot, y_test_onehot  = Test_data_onehot[:,:-1], Test_data_onehot[:,-1]
```
